#Top Senders and Recipients



SQL = """

SELECT sender As Sender , COUNT(*) As Number_of_Email FROM message
GROUP BY Sender
ORDER BY COUNT(*) DESC
LIMIT 12
"""
- SQl query used to extract, count, group and order frequent sender data from the message dataframe with a limit of 12 displayed information.

cur.execute(SQL)
- Executing the SQL query

F_Esenders_df = pd.read_sql(SQL,connect)
F_Esenders_df 
- Executing sql query by displaying frequent email sender form the dataframe. The 'sql,connect' is stored in the F_senders_df.

fig = plt.subplots(figsize=(20,10))
- Plotting bar chart for top sender using matplotlib and seaborn. The bar chart has a scale of 20,10.

sns.barplot(x= 'Number_of_Email', y= 'Sender',data= F_Esenders_df)
- The x-axis = Number_of_Email, the y-axis = sender and title = Top sender Email. This code is used to plot the bar chart.

plt.title('Top Sender Email')
plt.show()
- This codes displays the title of the graph.
 - It illustrates the bar chart for F_Esenders_df.
